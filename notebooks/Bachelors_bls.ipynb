```python
import pandas as pd
```

```python
unemp_bachelors = pd.read_csv('../data/unemp_bach2.csv')
df_bach= pd.DataFrame(unemp_bachelors)
df_bach
```

```python
df_bach=df_bach.drop(['Series ID', 'Period'], axis = 1) 
```

```python
df_bach
```

```python
type(df_bach['Label'][0])
```

```python
df_bach['Label'] = df_bach['Label'].str.replace(r'2019 ', '').astype(str)
```

```python
df_bach
```

```python
df_bach['Label'] = df_bach['Label'].str.replace(r'2020 ', '').astype(str)
```

```python
df_bach
```

```python
df_bach.set_axis(['Year', 'Month', 'Bachelors Degree'], axis=1, inplace=False)
```

```python
df_bach.to_csv('../output/highschool.csv', index=False)
```
