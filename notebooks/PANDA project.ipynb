```python
import pandas as pd
```

```python
unemp_bachelors = pd.read_csv('../data/unemp_bach2.csv')
df_bach= pd.DataFrame(unemp_bachelors)
df_bach
```

```python
df_bach=df_bach.drop(['Series ID', 'Year', 'Period'], axis = 1) 
```

```python
df_bach
```

```python
unemp_lhs = pd.read_csv('../data/unemp_lhs2.csv')
df_lhs= pd.DataFrame(unemp_lhs)
df_lhs
```

```python
df_lhs=df_lhs.drop(['Series ID', 'Year', 'Period'], axis = 1) 
```

```python
df_lhs
```

```python
merged_unemp = pd.merge(df_bach, df_lhs, on='Label')
```

```python
merged_unemp
```

```python
merged_unemp.set_axis(['Date', 'Bachelors', 'Less than Highschool'], axis=1, inplace=False)
```

```python
merged_unemp.to_csv('../output/merged_unemp.csv', index=False)
```
